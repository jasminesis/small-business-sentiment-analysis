```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
len(stop_words)
# Finding the number of stop words in each review
df = pd.read_csv('joined_reviews.csv')

#cleaning data
df['lowercase'] = df['text'].apply(lambda x: " ".join(word.lower() for word in x.split()))
df['punctuation'] = df['lowercase'].str.replace('[^\w\s]', '')
```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

df['stopwords'] = df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in  stop_words))
# Grabbing the top 30 words that occur in the highest frequency
pd.Series(" ".join(df['stopwords']).split()).value_counts()[:30]
#Removing other stop words that don't really contribute to calculations of the overall sentiment
other_stop_words = ['place', 'also', 'would', 
                    'got', 'ordered', 'one', 
                    'go', 'get','order','came',
                    'menu','restaurant','try',
                    'new york', 'us', 'come']

df['clean_review'] = df['stopwords'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stop_words))

pd.Series(" ".join(df['clean_review']).split()).value_counts()[:30]
```

```python
from textblob import Word
from textblob import TextBlob

df['lemmatized'] = df['clean_review'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
df['polarity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])

# Measuring content of personal opinion, emotion, or judgement
df['subjectivity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])

# Looking at general summary of the data so far
df.describe()
```

```python
df.sort_values(by='polarity')
df.to_csv('reviews_with_sentiment.csv')
```
